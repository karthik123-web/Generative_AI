# Task 4: ChatPromptTemplate System

```python
# Function to build chat prompt based on role
from langchain_core.prompts import ChatPromptTemplate
def build_chat_prompt(role, topic):

    # Define role behavior

    if role == "teacher":
        system_message = "You are a helpful teacher who explains topics clearly."

    elif role == "interviewer":
        system_message = "You are a technical interviewer who asks smart questions."

    elif role == "motivator":
        system_message = "You are a motivational speaker who encourages learning."

    else:
        system_message = "You are a helpful assistant."

    # Create Chat Template

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Explain {topic}")
    ])

    # Generate Prompt

    prompt = chat_template.format(
        topic=topic
    )

    return prompt


# Test Example

result = build_chat_prompt(
    role="teacher",
    topic="Neural Networks."
)

print("Output:")
print(result)
```

# Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):

    # Allowed values

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Check audience

    if audience not in valid_audience:
        print("Invalid audience. Default set to beginner.")
        audience = "beginner"

    # Check tone

    if tone not in valid_tone:
        print("Invalid tone. Default set to formal.")
        tone = "formal"

    return audience, tone


# Test the function

audience_test, tone_test = validate_inputs(
    audience="kids",     # invalid
    tone="friendly"      # invalid
)

print("\nValidated Values:")
print("Audience:", audience_test)
print("Tone:", tone_test)
```

# Task 6: Prompt Generator App 

```python
from langchain_core.prompts import PromptTemplate

def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate Inputs

    audience, tone = validate_inputs(audience, tone)

    # Step 2: Choose Template Based on Style

    if style == "teaching":

        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        )

        result = template.format(
            topic=topic,
            audience=audience,
            tone=tone
        )

    elif style == "interview":

        template = PromptTemplate(
            input_variables=["topic"],
            template="Ask 3 interview questions about {topic}"
        )

        result = template.format(
            topic=topic
        )

    elif style == "storytelling":

        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )

        result = template.format(
            topic=topic,
            audience=audience,
            tone=tone
        )

    else:

        template = PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic}"
        )

        result = template.format(
            topic=topic
        )

    return result


# Test the function

output = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

print("Output:")
print(output)
```

# Task 7: Template Reusability Test 

```python
from langchain_core.prompts import PromptTemplate

# Create ONE reusable template

reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# List of 5 different inputs

inputs_list = [
    ("AI", "beginner", "formal"),
    ("Python", "intermediate", "casual"),
    ("Deep Learning", "expert", "formal"),
    ("Data Science", "beginner", "fun"),
    ("Cloud Computing", "expert", "casual")
]

print("Task 7 Outputs:\n")

# Run template multiple times

for topic, audience, tone in inputs_list:

    # Validate inputs

    audience, tone = validate_inputs(
        audience,
        tone
    )

    # Generate prompt

    result = reusable_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

    print(result)
```
